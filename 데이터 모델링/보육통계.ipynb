```python
with open('dataset/어린이_교통사고_발생률.txt','r')as f:
    data = f.read()

d1 = data.replace('\n','\t').split('\t')
da1 = d1[9:18]

da2 = d1[18:]
year = da2[0::9]
year = pd.Series(year)
local = da2[1::9]
local = pd.Series(local)
child_a_val = da2[2::9]
child_a_val = pd.Series(child_a_val)
child_d_val = da2[3::9]
child_d_val = pd.Series(child_d_val)
child_i_val = da2[4::9]
child_i_val = pd.Series(child_i_val)
area_child_a = da2[5::9]
area_child_a = pd.Series(area_child_a)
area_child_i = da2[6::9]
area_child_i = pd.Series(area_child_i)
walk_child_d = da2[7::9]
walk_child_d = pd.Series(area_child_d)
walk_child_i = da2[8::9]
walk_child_i = pd.Series(walk_child_i)

col = pd.Series(da1)
df = pd.DataFrame([year,local,child_a_val,child_d_val,child_i_val,area_child_a,area_child_i,walk_child_d,walk_child_i])
df = df.transpose()
df.rename(col,axis=1,inplace=True)

df.columns = ['기간','지역','어린이 교통사고 발생건수','어린이 교통사고 사망자수','어린이 교통사고 부상자수','보호구역내 어린이 사고 발생건수','보호구역내 어린이 사고 부상자수','어린이 보행 사망자수','어린이 보행 부상자수']
df
# df.to_csv('dataset/지역별 어린이 사고율.csv')
# 어린이 인구 대비 비율로 다시
```
