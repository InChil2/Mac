```python
df2.fillna(0, inplace = True)
def classify(x):
    if x > df2['국공립_충족률'].mean():
        return 1
    else: return 0
df2['정원충족률_평균이상'] = df2['국공립_충족률'].apply(classify)
```

```python
df3 = df2[['만0~2세_국공립_보육수요', '만0~2세_인구수', '출생_계', '평균물건금액/평균건물면적', '합계_시설수', '전반적인_삶의_만족도',
       '지역_전반적인_생활_만족도', '이웃_알고지냄', '이웃_이야기', '이웃_도움', '이웃_참여', '임신/출산/육아_복지',
       '공포_밤에혼자', '공포_혼자골목길', '전반적인_사회안전', '환경_대기', '환경_수질', '환경_토양',
       '환경_소음/진동', '환경_녹지환경']]
df4 = df2[['정원충족률_평균이상', '만0~2세_국공립_보육수요', '만0~2세_인구수', '출생_계', '평균물건금액/평균건물면적',  '합계_시설수', '전반적인_사회안전', '환경_수질', '환경_토양']]
```

```python
# 예측모형

import numpy as np

# 독립변수, 종속변수 분리
y = df3['만0~2세_국공립_보육수요'] # 종속변수
X = df3.drop('만0~2세_국공립_보육수요', axis = 1) # 독립변수

# train data 와 test data로 구분
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,               #독립 변수 
                                                    y,               #종속 변수
                                                    test_size=0.2,   #검증 20%
                                                    random_state=12) #랜덤 추출 값 

# 단순회귀분석 모형 생성 및 평가
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lr = LinearRegression()
lr.fit(X_train,y_train)
y_preds = lr.predict(X_test)
mse = mean_squared_error(y_test,y_preds)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_preds)
print('MSE:{0:.4f}, RMSE:{1:.4f}, R2:{2:.4f}'.format(mse,rmse,r2))
```

```python
# 독립변수, 종속변수 분리
y = df4['정원충족률_평균이상'] # 종속변수
X = df4.drop('정원충족률_평균이상', axis = 1) # 독립변수

# train data 와 test data로 구분
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,               #독립 변수 
                                                    y,               #종속 변수
                                                    test_size=0.2,   #검증 20%
                                                    random_state=12) #랜덤 추출 값 
# 모델 학습 및 평가
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
accuracy_rf = accuracy_score(y_test, rf_pred).round(2)

lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
lr_pred = lr_model.predict(X_test)
accuracy_lr = accuracy_score(y_test,lr_pred).round(2)

print('rf 정확도:{}, lr 정확도:{}'.format(accuracy_rf,accuracy_lr))
```
